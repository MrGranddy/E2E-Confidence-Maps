```python
import os

from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt
```

```python
output_path = "G:/E2E-Confidence-Maps/shadow_segmentation_experiment_results"
selected_image = 1
```

```python
methods = os.listdir(output_path)
```

```python
results = {}

for method in methods:

    gt = os.path.join(output_path, method, f"gt_{selected_image}.png")
    pred = os.path.join(output_path, method, f"pred_{selected_image}.png")
    image = os.path.join(output_path, method, f"image_{selected_image}.png")

    results[method] = {"gt": Image.open(gt), "pred": Image.open(pred), "image": Image.open(image)}
    
```

```python
name_mappings = {
    "liver-acyclic": "$A_{CM}$",
    "liver-karamalis": "$K_{CM}$",
    "liver-mid": "$Mid_{CM}$",
    "liver-min": "$Min_{CM}$",
    "liver-no_cm": "No CM",
    "liver-reg0": "$M^{0}_{CM}$",
    "liver-reg01": "$M^{0.1}_{CM}$",
    "liver-reg001": "$M^{0.01}_{CM}$",
    "liver-reg0001": "$M^{0.001}_{CM}$",
    "liver-ultranerf": "$U_{CM}$",
}
```

```python
plt.figure(figsize=(8, 12))

plt.subplot(3, 4, 1)
plt.imshow(np.array(results["liver-acyclic"]["image"]), cmap="gray")
plt.axis("off")
plt.title("Image")

plt.subplot(3, 4, 2)
plt.imshow(np.array(results["liver-acyclic"]["gt"]), cmap="gray")
plt.axis("off")
plt.title("GT")

plt.subplot(3, 4, 3)
plt.imshow(np.array(results["liver-acyclic"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-acyclic"])

plt.subplot(3, 4, 4)
plt.imshow(np.array(results["liver-karamalis"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-karamalis"])

plt.subplot(3, 4, 5)
plt.imshow(np.array(results["liver-mid"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-mid"])

plt.subplot(3, 4, 6)
plt.imshow(np.array(results["liver-min"]["pred"]), cmap="gray") 
plt.axis("off")
plt.title(name_mappings["liver-min"])

plt.subplot(3, 4, 7)
plt.imshow(np.array(results["liver-no_cm"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-no_cm"])

plt.subplot(3, 4, 8)
plt.imshow(np.array(results["liver-reg0"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-reg0"])

plt.subplot(3, 4, 9)
plt.imshow(np.array(results["liver-reg01"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-reg01"])

plt.subplot(3, 4, 10)
plt.imshow(np.array(results["liver-reg001"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-reg001"])

plt.subplot(3, 4, 11)
plt.imshow(np.array(results["liver-reg0001"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-reg0001"])

plt.subplot(3, 4, 12)
plt.imshow(np.array(results["liver-ultranerf"]["pred"]), cmap="gray")
plt.axis("off")
plt.title(name_mappings["liver-ultranerf"])

plt.show()
```
