```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
array_path = "G:/masters-thesis/consistency_experiment_data/spine/right/cms_reg0.npy"
```

```python
array = np.load(array_path)
```

```python
random_indices = np.random.choice(array.shape[0], 8, replace=False)

fig, axs = plt.subplots(2, 4, figsize=(16, 8))

for i, ax in enumerate(axs.flat):
    ax.imshow(array[random_indices[i]])
    ax.axis("off")

plt.show()
```
